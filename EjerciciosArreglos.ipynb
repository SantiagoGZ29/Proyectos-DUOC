```python
import numpy as np
arr = np.random.randint(1,100,(3,3))

print(arr)

promedio = np.mean(arr)
print(promedio)

suma = np.sum(arr)
print(suma)

maximo = np.max(arr)
print(maximo)

minimo = np.min(arr)
print(minimo)

diagonal = np.diag(arr)
print(diagonal)







```
