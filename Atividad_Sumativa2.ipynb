```python
print("Bienvenido a KUMOO RACING")

patentes = []
marcas = []
modelos = []
años = []
tipos = []

registros = {}
#Menú de opciones

stop = 1
while stop == 1:

  print ("Tenemos las siguientes opciones: ")
  print ("1. Ingresar camioneta: ")
  print ("2. Registro de mantenimiento: ")
  print ("3. Consultar camioneta: ")
  print ("4. Salir: ")

  opcion = int(input("Ingrese una opcion(1,2,3 o 4)"))


  if opcion == 1: #Ingresar automovil
      print("Realice el ingreso de su vehículo: ")

      #Patente

      stop2 = 1
      while stop2 == 1:
        patente = (input("Ingrese la patente en formato AA1000 o BBBB10: ")).upper()
        def validar_patente(patente):
          if len(patente) != 6:
            return False
          if patente[:2].isalpha() and patente[2:].isdigit():
            return True
          elif patente[:4].isalpha() and patente[4:].isdigit():
            return True
          return False
        if validar_patente(patente):
          patentes.append(patente)
          registros[patente] = []
          print("patente válida")
          stop2 = 0
        else :
          print("patente inválida ingresar de nuevo en formato AA1000 o BBBB10")

      #año

      stop3 = 1
      while stop3 == 1:
        año = int(input("Ingrese el año de fabricación solo del 1975 al 2024: "))
        if año < 1975 or año > 2024:
          print("Año inválido")
        else:
          años.append(año)
          stop3 = 0
          print("Año válido")

      #Tipo de vehiculo

      stop4 = 1
      while stop4 == 1:
        print("Ingrese una opcion de tipo de vehiculo: ")
        print("1. Tracción simple (4x2) ")
        print("2. Tracción doble (4x4)")

        tipo = int(input("Ingrese una opción (1 o 2)"))

        if tipo == 1:
            tipos.append("Tracción_simple")
            print("Tipo de vehiculo válido")
            stop4 = 0
        elif tipo == 2:
            tipos.append("Tracción_doble")
            print("Tipo de vehiculo válido")
            stop4 = 0
        else:
            print("Opcion invalida escoja de nuevo")

      #Marca y modelo

      stop5 = 1
      while stop5 == 1:
        marca = input("Ingrese la marca del vehiculo: ")
        if marca == "":
          print("Ingrese una marca de vehículo")
        else:
          marcas.append(marca)
          stop5 = 0
      stop6 = 1
      while stop6 == 1:
        modelo = input("Ingrese el modelo del vehiculo: ")
        if modelo == "":
          print("Ingrese un modelo de vehículo")
        else:
          modelos.append(modelo)
          stop6 = 0
      print("Vehículo ingresado con éxito")

  elif opcion == 2:  #Registro de mantenimiento

      print("Registro de mantenimiento: ")
      patente = input("Ingrese la patente del vehiculo: ").upper()

      indice = patentes.index(patente)

      if patente in patentes:
        fecha = input("Ingrese la fecha de mantenimiento en formato dd/mm/aaaa ")
        observacion = input("Ingrese las observaciones: ")

        # Agregar el registro de mantenimiento a la lista asociada a la patente
        registros[patente].append((fecha, observacion))

        print("Registro ingresado con éxito")


      else :
        print("Patente no encontrada ingrese su vehículo en la opción 1.")
        stop = 1

  elif opcion == 3: #consultar automovil

      print("Consulta tú automovil")
      patente = input("Ingrese la patente del vehiculo: ").upper()
      indice = patentes.index(patente)
      if patente in patentes:

        print("Vehiculo encontrado: ")
        print("Patente:            ",(patentes[indice]))
        print("Año de fabricación: ",(años[indice]))
        print("Tipo de vehículo:   ",(tipos[indice]))
        print("Marca:              ",(marcas[indice]))
        print("Modelo:             ",(modelos[indice]))
        print("Registro:           ")
        for registro in registros[patente]:
          print(f"Fecha: {registro[0]}, Observaciones: {registro[1]}")


      else :
        print("Patente no encontrada ingrese su vehículo en la opción 1.")


  elif opcion == 4:
      print("Gracias por atenderse con nosotros")
      stop = 0
  else:
      print("Opcion invalida")
```
