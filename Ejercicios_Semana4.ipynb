```python
print("Bienvenido a la tienda de articulos sanitarios")

#Declaramos variables

mascarilla = 1000
guantes = 1000
delantal = 2000
amonio =  3000
total = 0
valorM = 0
valorG = 0
valorD = 0
valorA = 0
x = 0

print ("Tenemos un catalogo para ofrecerte \n mascarilla - $1.000 \n guantes - $1.000 \n delantal - $2.000 \n amonio - $3.000")

#Creamos el bucle de 4 para ingresar el pedido
while x < 4:
  compra = input("¿Que pruducto necesita?")
  if compra == "mascarilla":
    cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
    valorM = cantidad * mascarilla
  elif compra == "guantes":
    cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
    valorG = cantidad * guantes
  elif compra == "delantal":
    cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
    valorD = cantidad * delantal
  elif compra == "amonio":
    cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
    valorA = cantidad * amonio
  else:
    print("No tenemos ese producto")
  x += 1

#Sumamos el total
total = valorM + valorG + valorD + valorA

#Descuento
descuento = input("¿Tiene código de descuento si/no?")
if descuento == "si":
  porcentaje = int(input("Ingrese su % de descuento"))
  porcentaje = porcentaje / 100
  porcentaje = porcentaje * total
  total = total - porcentaje
elif descuento == "no":
  print("Pagaras el total de la compra")
else:
  print("Respuesta no válida")

#Imprimimos el resultado
print("Su compra es de: ")
print(f"mascarilla: ${valorM}")
print(f"guantes: ${valorG}")
print(f"delantal: ${valorD}")
print(f"amonio: ${valorA}")
print(f"Total: ${total}")
```

```python
azucar = 1200
agua = 600
aceite = 1500
arroz = 1250
fideos = 790
bebida = 1750
chocolate = 2500
pan = 1340

valorAzucar = 0
valorAgua = 0
valorAceite = 0
valorArroz = 0
valorFideos = 0
valorBebida = 0
valorChocolate = 0
valorPan = 0

total = 0

print("Bienvenido a la mi tienda tenemos el siguiente catálogo para usted: ")
print(" agua - $600 \n azucar - $1.200 \n aceite - $1.500 \n arroz - $1.250 \n fideos - $790 \n bebida - $1.750 \n chocolate - $2.500\n pan - $1.340 ")

while True:
    respuesta = input("¿Quieres comprar algún producto? (si/no): ")
    if respuesta.lower() == "no":
        break
    compra = input("¿Que producto necesita?")
    if compra == "azucar":
      cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
      valorAzucar = cantidad * azucar
    elif compra == "agua":
      cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
      valorAgua = cantidad * agua
    elif compra == "aceite":
      cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
      valorAceite = cantidad * aceite
    elif compra == "arroz":
      cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
      valorArroz = cantidad * arroz
    elif compra == "fideos":
      cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
      valorFideos = cantidad * fideos
    elif compra == "bebida":
      cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
      valorBebida = cantidad * bebida
    elif compra == "chocolate":
      cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
      valorChocolate = cantidad * chocolate
    elif compra == "pan":
      cantidad = int(input(f"¿Que cantidad de {compra} necesita?"))
      valorPan = cantidad * pan
    else:
      print("Producto no disponible")

total = valorAzucar + valorAgua + valorAceite + valorArroz + valorFideos + valorBebida + valorChocolate + valorPan


descuento = input("¿Es usted cliente preferencial si/no?")
if descuento == "si":
  print("Usted obtendrá un descuento del 25%")
  total = total * 0.75
else:
  print("Usted no obtendra un descuento ")

print(f"El total a pagar es de: ${total}")

efectivo = int(input("Ingrese su efectivo: "))
if efectivo < total:
  print("Dinero insuficiente")
else:
  cambio = efectivo - total
  print(f"Su cambio es de: ${cambio}")

print("Gracias por su compra")





```
