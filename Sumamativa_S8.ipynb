```python
import random as rd
import numpy as np

vehiculos = {}

#Funcion guardar:
def guardar ():

 while True:

    tipo = input("Ingrese el tipo de vehículo (Auto, Suv, Pickup, Van): ").upper()
    if tipo not in ["AUTO", "SUV", "PICKUP", "VAN"]:
        print("Tipo de vehículo no válido. Intente nuevamente.")
        continue

    patente = (input("Ingrese la patente en formato AA1000 o BBBB10: ")).upper()
    def validar_patente(patente):
      if len(patente) != 6:
        return False
      if patente[:2].isalpha() and patente[2:].isdigit():
        return True
      elif patente[:4].isalpha() and patente[4:].isdigit():
        return True
      return False
    if not validar_patente(patente):
        print("Patente no válida. Intente nuevamente.")
        continue


    marca = input("Ingrese la marca: ")
    if len(marca) < 2 and len(marca)<12 :
        print("La marca debe tener entre 2 y 12 caracteres.")
        continue

    modelo = input("Ingrese el modelo: ")

    precio = float(input("Ingrese el precio: "))
    if precio < 6880000:
        print("El precio no puede ser inferior a $6.880.000")
        continue

    fecha_registro = input("Ingrese la fecha de registro (DD-MM-AAAA): ")
    nombre_dueño = input("Ingrese el nombre del dueño: ")


    multas = []
    while True:
       agregar_multa = input("¿Desea agregar una multa? (si/no): ").lower()
       if agregar_multa.lower() != "si":
        break
       while True:
          fecha_multa = input("Ingrese la fecha de la multa (DD-MM-AAAA): ")
          monto_multa = rd.randint(2300, 5400)
          multas.append((monto_multa, fecha_multa))
          break

    Emision_contaminantes = input("Ingrese la emision de contaminantes(Cumple o No cumple): ")
    anotaciones_vigentes = input("Ingrese las anotaciones vigentes: ")
    if anotaciones_vigentes == "":
        anotaciones_vigentes = "No hay anotaciones "

    vehiculo = {
        "tipo": tipo,
        "patente": patente,
        "marca": marca,
        "modelo": modelo,
        "precio": precio,
        "multas": multas,
        "fecha_registro": fecha_registro,
        "nombre_dueño": nombre_dueño,
        "Emision_contaminantes": Emision_contaminantes,
        "Anotaciones_vigentes": anotaciones_vigentes
    }

    vehiculos[patente] = vehiculo

    continuar = input("¿Desea ingresar otro vehículo? (si/no): ")
    if continuar.lower() != "si":
        break

#Funcion buscar:

def buscar():
  patente = input("Ingrese la patente del vehículo que desea buscar: ").upper()
  if patente in vehiculos:
    print("Vehículo encontrado:")
    vehiculo = vehiculos[patente]
    print("Tipo de vehiculo: ",vehiculo["tipo"])
    print("Patente: ",vehiculo["patente"])
    print("Marca: ",vehiculo["marca"])
    print("Modelo: ",vehiculo["modelo"])
    print("Precio: ",vehiculo["precio"])
    print("Multas: ",vehiculo["multas"])
    print("Fecha de registro: ",vehiculo["fecha_registro"])
    print("Nombre del dueño: ",vehiculo["nombre_dueño"])

  else:
    print("Vehículo no encontrado.")

# Funcion de imprimir certificado:
def imprimir_certificado():
  patente = input("Ingrese la patente del vehículo que desea buscar: ").upper()
  if patente in vehiculos:

    vehiculo = vehiculos[patente]
    while True:
        print("1. Emisión de contaminantes \n2. Anotaciones vigentes \n3. Multas \n4. Salir")
        opcion = int(input("Ingrese una opción "))
        if opcion == 1:
          print("Certificado de emisión de contaminantes:")
          print("Emision de contaminantes: ", vehiculo["Emision_contaminantes"])
          print("Patente: ", vehiculo["patente"])
          print("Dueño: ", vehiculo["nombre_dueño"])
        elif opcion == 2:
          print("Certificado de anotaciones vigentes:")
          print("Anotaciones vigentes: ", vehiculo["Anotaciones_vigentes"])
          print("Patente: ", vehiculo["patente"])
          print("Dueño: ", vehiculo["nombre_dueño"])
        elif opcion == 3:
          print("Certificado de multas:")
          for monto_multa, fecha_multa in vehiculo["multas"]:
            print("Monto de la multa: ", monto_multa)
            print("Fecha de la multa: ", fecha_multa)
        elif opcion == 4:
          break
        elif opcion < 1 or opcion > 4:
          print("Opción no válida. Intente nuevamente.")
  elif patente not in vehiculos:
    print("Vehículo no encontrado.")

#Funcion salir

def salir():
  print("Gracias por usar la automotora Ruta 40")
  print("Santiago Gallego Zamora, versión 1.0")

print("Bienvenido a la automotora Ruta 40")

opcion = 0

while opcion != 4:
  print("1. Guardar \n2. Buscar \n3. Impresion de certificados \n4. Salir")
  opcion = int(input("Ingrese una opción "))

  if opcion == 1:
    guardar()
  elif opcion == 2:
    buscar()
  elif opcion == 3:
    imprimir_certificado()
  elif opcion == 4:
    salir()
    break
  elif opcion < 1 or opcion > 4:
    print("Opción no válida. Intente nuevamente.")













```
