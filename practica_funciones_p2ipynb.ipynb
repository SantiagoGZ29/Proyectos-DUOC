```python
def es_primo(numero):
  if numero <= 1:
    return False
  for i in range(2, int(numero**0.5) + 1):
    if numero % i == 0:
      return False
  return True

def factorial(numero):
  if numero == 0:
    return 1
  else:
    return numero * factorial(numero - 1)

def es_palindrome(cadena):
  return cadena == cadena[::-1]


print("Binvenido a nuestro men de opciones: ")

opcion = 0

while opcion != 4:
  print("1. Nmero primo \n2. Numero factorial \n3. palindrome \n4. Salir ")
  opcion = int(input("Ingrese una opcion del men "))

  if opcion == 1:
    numero = int(input("Ingrese un nmero entre 1 y 15 para comprobar si es primo o no: "))
    if es_primo(numero):
      print(numero, "es un nmero primo")
    else:
      print(numero, "no es un nmero primo")
  if opcion == 2:
    numero = int(input("Ingrese un nmero para calcular su factorial: "))
    print("El factorial de", numero, "es", factorial(numero))
  if opcion == 3:
    cadena = input("Ingrese una cadena para comprobar si es un palindrome o no: ")
    if es_palindrome(cadena):
      print(cadena, "es un palindrome")
    else:
      print(cadena, "no es un palindrome")
  if opcion == 4:
    break
  else:
    print("Opcin invlida")
```
