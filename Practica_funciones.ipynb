```python
def imc (peso,altura):
  imc = peso / (altura **2)
  print("T IMC es de: ", imc)
  if imc < 18.5:
    print("Estas bajo de peso")
  if imc >= 18.5 and imc <= 24.9:
    print("Estas en un peso normal")
  if imc >= 25 and imc <= 29.9:
    print("Estas en sobrepeso")
  if imc >= 30 and imc <= 34.9:
    print("Estas en obesidad grado 1")
  if imc >= 35 and imc <= 39.9:
    print("Estas en obesidad grado 2")
  if imc >= 40:
    print("Estas en obesidad grado 3")

def desc (precio_inicial,descuento):
  desc  = precio_inicial * descuento
  precio_final = precio_inicial - desc

  print("El precio inicial es de: $", precio_inicial)
  print("El descuento es de: $",desc)
  print("El precio final con descuento es de: $", precio_final)
  return precio_final


def calcular_iva(precio):
  iva = precio * 0.19
  return iva


print("Bienvenido a nuestro men")

opcion = 0

while opcion !=4:
    print("1. Calcular IVA\n2. Calcular descuento\n3. Calcular IMC\n4. Salir del men")
    opcion = int(input("Ingrese una opcion"))

    if opcion == 1:
        precio = float(input("Ingrese el precio del producto "))
        print("El iva es de: $",calcular_iva(precio))

    if opcion == 2:
      precio_inicial = int(input("Ingrese el precio inicial"))
      descuento = float(input("Ingrese el procentaje de descuento")) /100
      desc(precio_inicial,descuento)

    if opcion == 3:
      peso= float(input("Ingrese su peso en KG "))
      altura= float(input("Ingrese su altura en metros"))
      imc(peso,altura)

    if opcion == 4:
      print("Gracias por usar nuestro men")
      break

```
