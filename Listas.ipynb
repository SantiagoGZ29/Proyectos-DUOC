# **Operaciones con listas**

Las listas permiten almacenar los datos que el algoritmo necesita para funcionar.

Existen varios m√©todos u operaciones para construir y modificar una lista, entre ellas  append, extend, insert, remove, pop, reverse, sort, e iterar listas.

A continuaci√≥n, se revisar√°n cada uno de ellos.

**Append(< obj >)**

El m√©todo append(), permite agregar elementos al final de la lista.

```python
lista = [1,2]
print(lista)
lista.append(3)
print(lista)
```

**Extend(< iterable >)**

El m√©todo extend(), permite a√±adir una lista a una lista inicial.

```python
lista = [1,2]
print(lista)
lista.extend([3,4])
print(lista)
```

**Insert(< index >, < obj >)**

El m√©todo insert(), permite a√±adir un elemento en una posici√≥n determinada.

```python
lista = [1,3]
print(lista)
lista.insert(1,2) #La primnera es posicion y la segunda lo que quieres ingresar
print(lista)
```

**Remove(< obj >)**

El m√©todo remove(), permite recibir como argumento un objeto y borrarlo de la lista.

```python
lista = [0,2,3]
print(lista)
lista.remove(0) #No es la posicion si no el elemento
print(lista)
```

**Pop()**

El m√©todo pop permite eliminar elementos de dos formas:

```python
lista = [1,2,3]
lista.pop() #Elimina el √∫ltimo elemento
print(lista)
lista = [1,2,3]
lista.pop(0) #Elimina el elemento en la posici√≥n 0
print(lista)
```

**Reverse()**

El m√©todo reverse(), permite invertir el orden de la lista

```python
lista = [1,2,3]
print(lista)
lista.reverse()
print(lista)
```

sort()

El m√©todo sort(), permite ordenar el orden de una lista

```python
lista = [6,2,7,1]
print(lista)
lista.sort() #De menor a mayor
print(lista)

lista = [6,2,7,1]
lista.sort(reverse=True) #De mayor a menor
print(lista)
```

**Iterar listas**

Otro de los procedimientos que se pueden realizar es iterar listas. A continuaci√≥n, se muestra un ejemplo:

```python
lista = [1,2,3]
for i in lista:
  print(i)
```

**M√©todos de cadenas en Python**
Estos m√©todos son muy √∫tiles para manipular cadenas de texto en Python.

**1. upper()**

Funci√≥n: Convierte todas las letras de una cadena a may√∫sculas.

```python
texto = "Hola, mundo"
texto_mayusculas = texto.upper()
print(texto_mayusculas)  # Imprime: HOLA, MUNDO!
```

**2. lower()**

Funci√≥n: Convierte todas las letras de una cadena a min√∫sculas.
Ejemplo:

```python
texto = "ESTO ES UN EJEMPLO"
texto_minusculas = texto.lower()
print(texto_minusculas)  # Imprime: esto es un ejemplo
```

**3. find()**

Funci√≥n: Busca una subcadena dentro de una cadena y devuelve el √≠ndice de la primera aparici√≥n. Si no encuentra la subcadena, devuelve -1.
Ejemplo:

```python
texto = "Estoy aprendiendo Python"
posicion = texto.find("Python")
print(posicion)  # Imprime: 18 (la palabra "Python" comienza en el √≠ndice 18)
```

**4. replace()**

Funci√≥n: Reemplaza todas las ocurrencias de una subcadena por otra.
Ejemplo:

```python
texto = "Hola, mundo! Hola otra vez."
nuevo_texto = texto.replace("Hola", "Adi√≥s")
print(nuevo_texto)  # Imprime: Adi√≥s, mundo! Adi√≥s otra vez.
```
