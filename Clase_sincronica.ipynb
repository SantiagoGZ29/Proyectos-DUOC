```python
print("Bienvenido, conoce tú tipo impositivo")
renta = int(input("¿Cuál es tú renta anual?: "))
impuesto = 0
if renta < 10000:
    print("Debes pagar el 5% de impuesto")
    impuesto = renta * 0.05
elif renta >= 10000 and renta < 20000:
    print("Debes pagar el 15% de impuesto")
    impuesto = renta * 0.15
elif renta >= 20000 and renta < 35000:
    print("Debes pagar el 20% de impuesto")
    impuesto = renta * 0.20
elif renta >= 35000 and renta < 60000:
    print("Debes pagar el 30% de impuesto")
    impuesto = renta * 0.30
elif renta >= 60000:
    print("Debes pagar el 45% de impuesto")
    impuesto = renta * 0.45
print("Tu impuesto es de: ", impuesto,)


```
