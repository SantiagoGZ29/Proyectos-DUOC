```python
print("Bienvenido al cineDuoc")

#definimos las variables

boleta = 0
valorentrada = 0
valorpalomitas = 0
valorbebidas = 0

#Verificamos si el cliente pertenece o no al DUOC

cliente = input("Pertenece al DUOC (si/no)?")
if cliente == "si":
    pertenece = input("Muestra t identificacin ")
    if pertenece == "verdadero":
      print("Bienvenido al cineDuoc perteneces al DUOC")
else:
  print("Bienvenido al cineDuoc no perteneces al DUOC")

#Pedimos la entrada que desea el cliente

entrada = input("Que entrada desea?\n1. estreno  $4.800\n2. normal  $2.900")
if entrada == "1":
  valorentrada =  4800
elif entrada == "2":
  valorentrada = 2900
numentrada = int(input("Ingrese el nmero de entradas?"))
if cliente == "si" and pertenece =="verdadero":
  valorentrada = numentrada * valorentrada * 0.70
else:
  valorentrada = valorentrada * numentrada

boleta = boleta + valorentrada

#Pedimos palomitas y bebidas

palomitas = input("Desea agregar palomitas de maz a su pedido?")
if palomitas == "si":
  promopalomitas = input("Selecione alguna promocin: \n1. Palomitas pequeas  $2.500 \n2. Palomitas medianas  $4.500 \n3. Palomitas grandes  $7.800 ")
  if promopalomitas == "1":
    valorpalomitas = valorpalomitas + 2500
  elif promopalomitas == "2":
    valorpalomitas = valorpalomitas + 4500
  elif promopalomitas == "3":
    valorpalomitas = valorpalomitas + 7800
  cantidadpp = int(input("Ingrese la cantidad de palomitas"))
  valorpalomitas = valorpalomitas * cantidadpp
  boleta = boleta + valorpalomitas
elif palomitas == "no":
  print("No agregar palomitas")
else:
  print("Valor mal ingresado")

#Bebidas

bebidas = input("Desea agregar bebidas a su pedido?")
if bebidas == "si":
  promobebidas = input("Selecione alguna promocin: \n1. Bebida pequea  $1.000  \n2. Bebida mediana  $1.250  \n3. Bebida grande  $2.000 ")
  if promobebidas == "1":
    valorbebidas = valorbebidas + 1000
  elif promobebidas == "2":
    valorbebidas = valorbebidas + 1250
  elif promobebidas == "3":
    valorbebidas = valorbebidas + 2000
  cantidadpb = int(input("Ingrese la cantidad de bebidas"))
  valorbebidas = valorbebidas * cantidadpb
  boleta = boleta + valorbebidas
elif bebidas == "no":
  print("No agregar bebidas")
else:
  print("Valor mal ingresados")

#mostramos el total a pagar
print(f"Su total a pagar es de: \n Entradas: ${valorentrada}\n Palomitas ${valorpalomitas} \n Bebidas ${valorbebidas} \n TOTAL: ${boleta}")

```
