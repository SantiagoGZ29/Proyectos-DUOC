```python
print("Bienvenido al cineDuoc")

#definimos las variables

boleta = 0
valorentrada = 0
valorpalomitas = 0
valorbebidas = 0

#Verificamos si el cliente pertenece o no al DUOC

cliente = input("¿Pertenece al DUOC?")
if cliente == "si":
  respuesta = input("estudiante o funcionario: ")
  if respuesta == "estudiante" or respuesta == "funcionario":
    pertenece = input("Muestra tú identificación ")
    if pertenece == "v":
      print("Bienvenido al cineDuoc perteneces al DUOC")
else:
  print("Bienvenido al cineDuoc no perteneces al DUOC")

#Pedimos la entrada que desea el cliente

entrada = input("¿Que entrada desea?\nestreno → $4.800\nnormal → $2.900 ")
if entrada == "estreno":
  valorentrada =  4800
elif entrada == "normal":
  valorentrada = 2900
numentrada = int(input("¿Ingrese el número de entradas?"))
if cliente == "si":
  valorentrada = valorentrada * numentrada * 0.70
else:
  valorentrada = valorentrada * numentrada

boleta = boleta + valorentrada

#Pedimos palomitas y bebidas

palomitas = input("¿Desea agregar palomitas de maíz a su pedido?")
if palomitas == "si":
  promopalomitas = input("Selecione alguna promoción: \nPalomitas pequeñas → $2.500 \nPalomitas medianas → $4.500 \nPalomitas grandes → $7.800 ")
  if promopalomitas == "pequeñas":
    valorpalomitas = valorpalomitas + 2500
  elif promopalomitas == "medianas":
    valorpalomitas = valorpalomitas + 4500
  elif promopalomitas == "grandes":
    valorpalomitas = valorpalomitas + 7800
  cantidadpp = int(input("Ingrese la cantidad de palomitas"))
  valorpalomitas = valorpalomitas * cantidadpp
  boleta = boleta + valorpalomitas
else:
  print("No agregará palomitas")

#Bebidas

bebidas = input("¿Desea agregar bebidas a su pedido?")
if bebidas == "si":
  promobebidas = input("Selecione alguna promoción: \nBebida pequeña → $1.000    \nBebida mediana → $1.250    \nBebida grande → $2.000 ")
  if promobebidas == "pequeñas":
    valorbebidas = valorbebidas + 1000
  elif promobebidas == "medianas":
    valorbebidas = valorbebidas + 1250
  elif promobebidas == "grandes":
    valorbebidas = valorbebidas + 2000
  cantidadpb = int(input("Ingrese la cantidad de bebidas"))
  valorbebidas = valorbebidas * cantidadpb
  boleta = boleta + valorbebidas
else:
  print("No agregará bebidas")

#mostramos el total a pagar
print(f"Su total a pagar es de: \n Entradas: ${valorentrada}\n Palomitas ${valorpalomitas} \n Bebidas ${valorbebidas} \n TOTAL: ${boleta}")

```
