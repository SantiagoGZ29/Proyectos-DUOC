```python
sw = 1
while sw == 1:
  print("1. Opcion 1")
  print("2. Opcion 2")
  print("3. Salir")
  op = int(input("Ingrese una opcion: "))
  try:
    if (op < 0 and op < 4):
      if op == 1:
        print("Selecionaste la opcion 1")
        continu=int(input("Desea salir 1=Si 2=No: "))
        if continu==1:
          sw=0
      if op == 2:
        print("Selecionaste la opcion 2")
        continu=int(input("Desea salir 1=Si 2=No: "))
        if continu==1:
          print("Adios")
      if op == 3:
        print("Usted desea salir del menú")
        sw=0
  except:
    print("Ingreso erroneo")
```
